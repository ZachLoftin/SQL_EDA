Now I need to test the hypothesis: "The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

Null Hypothesis (H0):
The average duration of rides from the Loop to O'Hare International Airport is the same on rainy Saturdays as on non-rainy Saturdays.

Alternative Hypothesis (H1):
The average duration of rides from the Loop to O'Hare International Airport is different on rainy Saturdays compared to non-rainy Saturdays.

Significance leve (alpha): 0.05 (5%)

I chose the null hypothesis because it is what I am going to be testing against, and the alternative hypothesis is what I am aiming to prove.

I'll be filtering the data to have two different data sets, one showing non rainy saturdays and one showing rainy Saturdays. This will make determining if there is a significant difference in the ride durations between rainy and non-rainy Saturdays simpler.

<div class="alert alert-success"; style="border-left: 7px solid green">
<b>✅ Reviewer's comment, v. 1</b> 
    
Hypotheses are formulated</div>

```python
#filter data to show rainy saturdays and non-rainy saturdays
sat_rides = rides_df[rides_df['start_ts'].dt.weekday == 5]
rainy_sat = sat_rides[sat_rides['weather_conditions'] == 'Bad']
non_rainy_sat = sat_rides[sat_rides['weather_conditions'] == 'Good']

#perform t-test
results = st.ttest_ind(non_rainy_sat['duration_seconds'], rainy_sat['duration_seconds'])

#make statistical significance level
alpha = 0.05

print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("Reject the null hypothesis: There is a significant difference in ride durations on rainy and non-rainy Saturdays.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in ride durations on rainy and non-rainy Saturdays.")
```

The final conclusion is that the ride durations bewteen rainy Saturdays and non-rainy Saturdays are significantly different.
